Assignment 3 - Working with Matplotlib
This script utilizes Matplotlib to generate line plots, bar graphs, pie charts, and histograms.

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()

students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = (0.1, 0, 0, 0)

plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title("Revenue Distribution by Region")
plt.show()

random_data = np.random.randint(1, 101, 1000)

plt.hist(random_data, bins=10, color='purple', edgecolor='black')
plt.title("Frequency Distribution of Random Integers")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```
