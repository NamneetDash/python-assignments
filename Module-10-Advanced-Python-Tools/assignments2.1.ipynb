Assignment 1 - Working with NumPy
In this assignment, we perform various operations using the NumPy library, including array creation, statistical calculations, reshaping, and matrix operations.

```python
import numpy as np


arr_1d = np.arange(1, 21)


print("Sum:", np.sum(arr_1d))
print("Mean:", np.mean(arr_1d))
print("Median:", np.median(arr_1d))
print("Standard Deviation:", np.std(arr_1d))


indices = np.where(arr_1d > 10)
print("Indices of elements > 10:", indices)


arr_2d = np.arange(1, 17).reshape(4, 4)


print("\n4x4 Array:\n", arr_2d)


print("Transpose:\n", arr_2d.T)


print("Row-wise sums:", np.sum(arr_2d, axis=1))
print("Column-wise sums:", np.sum(arr_2d, axis=0))


array1 = np.random.randint(1, 21, (3, 3))
array2 = np.random.randint(1, 21, (3, 3))

print("\nArray 1:\n", array1)
print("Array 2:\n", array2)


print("Addition:\n", array1 + array2)
print("Subtraction:\n", array1 - array2)
print("Multiplication:\n", array1 * array2)


print("Dot Product:\n", np.dot(array1, array2))


arr_size_12 = np.arange(12)
reshaped_arr = arr_size_12.reshape(3, 4)
sliced_arr = reshaped_arr[:2, -2:]

print("\nReshaped 3x4 Array:\n", reshaped_arr)
print("Sliced Array (First 2 rows, last 2 columns):\n", sliced_arr)
```
